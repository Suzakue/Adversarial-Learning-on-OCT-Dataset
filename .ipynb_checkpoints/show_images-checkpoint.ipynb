```python
import random
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

DATA_DIRECTORY = '/home/hejiawen/pytorch/ZRRP/AI-OCT/data/cutted_data/'
DATA_LIST_PATH = '/home/hejiawen/pytorch/ZRRP/AI-OCT/data/cutted_data_path/val_images.txt'
LABEL_LIST_PATH = '/home/hejiawen/pytorch/ZRRP/AI-OCT/data/cutted_data_path/val_labels.txt'
pred_dir = 'results_OCT_0607/'
NUM = 10

image_list = list()
label_list = list()

with open(DATA_LIST_PATH, 'r') as f:
    for line in f.readlines():
        line = line.strip()
        image_list.append(line)

with open(LABEL_LIST_PATH, 'r') as f:
    for line in f.readlines():
        line = line.strip()
        label_list.append(line)

if len(image_list) != len(label_list):
    print('image_list:', len(image_list))
    print('label_list:', len(label_list))

rand_list = [i for i in range(len(image_list))]
slice_list = random.sample(rand_list, NUM)

for i in range(NUM):
    idex = slice_list[i]
    line = image_list[idex]
    image_path = DATA_DIRECTORY + line
    label_path = DATA_DIRECTORY + label_list[idex]
        
    line = line.split('/')
    name = line[-2] + '_' + line[-1]
    pred_path = pred_dir + name
    
    image = Image.open(image_path)
    label = Image.open(label_path)
    # label = cv2.cvtColor(label, cv2.COLOR_GRAY2BGR)
    pred = Image.open(pred_path)
    
    plt.suptitle('image_' + str(idex))
    plt.subplot(1,3,1)
    plt.title('image')
    plt.imshow(image)
    
    plt.subplot(1,3,2)
    plt.title('label')
    plt.imshow(label,cmap='gray')
    
    plt.subplot(1,3,3)
    plt.title('pred')
    plt.imshow(pred)
    
    plt.show()
    
```
